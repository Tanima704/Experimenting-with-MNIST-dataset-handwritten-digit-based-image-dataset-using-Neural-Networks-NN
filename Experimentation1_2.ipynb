## <font color='blue'> MLP with scikit-learn: Explore outputs for the MNIST dataset </font>




```python
# Let's first load the MNIST data set. 

from sklearn.datasets import fetch_openml
from sklearn.model_selection import train_test_split


X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
y = y.astype(int)
X = ((X / 255.) - .5) * 2
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=10000, random_state=123, stratify=y)
```

```python
print(X_train)
```

<br> Find a setting for *batch_size* and *max_iter*, so that the subsequent fit function **converges**. 





```python


mlp_mnist = MLPClassifier(verbose = True, hidden_layer_sizes=(50,), batch_size = 10, \
                          max_iter=228, solver='sgd', activation='logistic',\
                          learning_rate = 'constant', learning_rate_init = 0.001, random_state=1)

```

```python


mlp_mnist.fit(X_train,y_train)
#
```

```python


#Hence for the setting of batch_size = 10,   max_iter=228 , the convergence is acheieved
```

```python
predictions = mlp_mnist.predict(X_test)
probability_predictions = mlp_mnist.predict_proba(X_test)
```

Write a short function that outputs the **second best** prediction for each point, i.e. the label which gets the second highest probability in the softmax output. 

```python


def second_best_prediction(p1):
    a=[]
    for p in p1:
        a.append(np.argsort(np.array(p))[-2])
    return(a)
print(second_best_prediction(probability_predictions))
```

For each data point misclassified by the model, find if the second best prediction is actually the correct label. Calculate the percentage of missclassified points, that are correctly classified based on the second best prediction. 

```python


a=second_best_prediction(probability_predictions)
b=y_test.values
missclassify=[]
original_for_missclassify=[]
for i in range(0,len(b)):
    if(a[i]==b[i]):
        missclassify.append(a[i])
        original_for_missclassify.append(b[i])


print('Percentage of misclassified points = ',len(missclassify)/len(y_test)*100)
```

```python


#This shows that only 2.19% of the points are missclassified and hence the highest probability level 
#is truly matching the original label in more than 97% of the scenarios
```

## <font color='blue'>Working with warm start for focused training



MLP classifiers in scikit-learn have an option for so-called 'warm start'.

Give an example of warm start. Specifically
train the network for $k$ epochs, and then using warm start
continue the training for another $k$ epochs (4 points).
Here $k$ is a parameter of my choice.  





```python
import time
```

```python
 

start = time.time()

mlp_test = MLPClassifier(verbose = True, hidden_layer_sizes=(50,), batch_size = 10, \
                          max_iter=5, solver='sgd', activation='logistic',\
                          learning_rate = 'constant', learning_rate_init = 0.001, random_state=1,warm_start=True)

mlp_test.fit(X_train,y_train)
end = time.time()
print(end - start)
#Here I have initiated MLP classifiers with warm_start = True. For 5 epochs, it took about 72 seconds.
```

```python


start = time.time()
mlp_test.fit(X_train,y_train)
end = time.time()
print(end - start)
#Continuing on the same previously initiated MLP classifier with warm_start = True, 
#while I again started a new training, it started from iteration 6. Next 5 iterations look 20 seconds less also.
# this is the beauty of warm_start where the training continues from where it ended last time. 
#This saves a lot of time and computing resource as same fitting is not done again and again. 
```


The training set usally remains the same after a warm start. Here, I will use a very similar bit not identical dataset after the warm start:
<br>
<br>
Randomly divide the MNIST dataset into three part: train1 (50000), train2(10000), test (10000). Make sure that all number classes are presented in both train1, and train2. [4 points]

Use warm start and train the model using the train1 dataset for $2k$ epochs. Then, using warm start, continute the training with train2 dataset for another $2k$ epochs. [4 points]
<br>
<br>
Here $k$ is the same parameter used above. Notice that each point
is considered $2k$ times in this training, exactly as in part 0. 
<br>
<br>
In effect this process of training tries spend more time learning half of the training set, and then moves to the second part of the training set to spend equal time. 


```python


start = time.time()

mlp_test = MLPClassifier(verbose = True, hidden_layer_sizes=(50,), batch_size = 10, \
                          max_iter=10, solver='sgd', activation='logistic',\
                          learning_rate = 'constant', learning_rate_init = 0.001, random_state=1,warm_start=True)

mlp_test.fit(X_train_1_new,y_train_1_new)
end = time.time()
print(end - start)

```

```python


start = time.time()
mlp_test.fit(X_train_2_new,y_train_2_new)
end = time.time()
print(end - start)
```

```python


# For 10 epochs withsame training dataset - it took 145.25 secs .
# For next 10 epochs with warm_start and  10000  records - it took only 67 seconds and also the accuracy improved 
# from 0.1854 to 0.1561. So this strategy works well for retraining models faster without the need to start freshly 
# with the full dataset. with weights and biases initialized to their existing values retraining happens faster
```

##  <font color='blue'> How Powerful are Neural Networks? 

There is a lot of hype around Neural Networks, and their ability to solver problems. This is of course justified, but I should always be aware of their limitations. 


Recall that every row in the matrix $X$  above comes from an image. Every entry of a row corresponds to a (grayscale) pixel. In particular, the first colum of $X$ contains the grayscale values of the first pixels of the images in the data set. 
<br>
<br>
I will now 'hide' the label of the image, inside the image itself. 
Here is one way to do that: Replace the top-left pixel of the image (which is a value from 0 to 255) with the label for that image. For example, if the image
contains number 3, then the first pixel of the image will now be set to 3. (In fact, let's set this to 3/255 because I already divided all pixel values by 255).

Create a new dataset $X1$, where each row is the same as the corresponding row of $X$, except in the first column, which contains the label of the corresponding image, divided by 255. Then split
the set $X1$ into a training set and a test set, as I did with $X$. 




```python


X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
y = y.astype(int)
X = ((X / 255.) - .5) * 2
print(X)
```

```python


X1=X
X1.pixel1=y/255
print(X1)
```

```python


X_train, X_test, y_train, y_test = train_test_split( X1, y, test_size=10000, random_state=123, stratify=y)
```

With the label of the image 'planted' in the image itself, a Neural Network can achieve accuracy 100%. This is because all the weights connecting to pixels $2...28^2$ can be set to 0, and so effectively the only feature that will be taken into account is the first pixel. The question is whether an MLP can indeed 'discover' this weighting via training.
<br>
<br>
Use an MLP classifier  with one hidden layer of any size. Train it on $X'$ and report the accuracy. Does it reach 100\%? Is the accuracy better relative to training with the original input $X$? 




```python

mlp_mnist_for_planted = MLPClassifier(verbose = True, hidden_layer_sizes=(5,), batch_size = 10, \
                          max_iter=20, solver='sgd', activation='logistic',\
                          learning_rate = 'constant', learning_rate_init = 0.001, random_state=1)

mlp_mnist_for_planted.fit(X_train,y_train)
```

```python

from sklearn.metrics import confusion_matrix
y_pred = mlp_mnist_for_planted.predict(X_test)
cm = confusion_matrix(y_pred, y_test)
print( cm)

from sklearn.metrics import accuracy_score
accuracy_score(y_pred, y_test)

#Below is the confusion Matrix and accuracy score is 91.47%. SO even if the original labels are planted 
#in the first pixel values, still 100% accuracy is not reached
```

```python


#Here the below code represents the accuracy using original data (no implanted label within the image) with same hyperparameters.
#This will help to compare the accuracy with data having the label implanted at the top left pixel.

X, y = fetch_openml('mnist_784', version=1, return_X_y=True)
y = y.astype(int)
X = ((X / 255.) - .5) * 2
X_train_orig, X_test_orig, y_train_orig, y_test_orig = train_test_split( X, y, test_size=10000, random_state=123, stratify=y)
mlp_mnist_original = MLPClassifier(verbose = True, hidden_layer_sizes=(5,), batch_size = 10, \
                          max_iter=20, solver='sgd', activation='logistic',\
                          learning_rate = 'constant', learning_rate_init = 0.001, random_state=1)

mlp_mnist_original.fit(X_train_orig,y_train_orig)

from sklearn.metrics import confusion_matrix
y_pred_orig = mlp_mnist_original.predict(X_test_orig)
cm_orig = confusion_matrix(y_pred_orig, y_test_orig)
print( cm_orig)

from sklearn.metrics import accuracy_score
accuracy_score(y_pred_orig, y_test_orig)

```

```python
#So from above two (one with label implanted within image and another with original data) confusion Matrices and 
# accuracy scores , I can conclude that original data based prediction accuracy (accuracy - 85.93% ) is lower 
#than the data having original label planted in pixel1 (accuracy - 91.43%). Hence MLP CLassifier is not able to predict 
# with 100% accuracy (but obviously predicted with better accuracy than original data) even though the Label is 
#implanted in the image itself.
```
